作业四 文本纠错  
任务描述：给出三个文件，词典库vocab.txt（任何不在词典中出现的词都认为是拼写错误），spell-errors.txt（给出正确单词与常见的拼写错误）与测试数据testdata.txt（1000个存在拼写错误的句子），利用bigram语言模型与Noisy Channel Model进行文本纠错训练。  
步骤（每一步注释标注）：  
一、读取词典库vocab.txt；  
二、对于词库里没有的词，取编辑距离为1和2，生成候选词集合；  
三、读取spell-errors.txt，计算概率p(错误的单词|正确的单词)；  
四、选用nltk语料库movie_reviews，构建bigram语言模型；  
五、读取testdata.txt，找出拼错词，生成候选词，并计算每个候选词的概率；  
六、选出概率最高的候选词，作为修改词。  
注：为便于计算，可采用似然对数，将概率乘积转化为求和。注意平滑。  
输出：对每个句子，输出句子编号，拼错词，修改词（若没有则输出False），候选词列表及概率。最后输出总拼错词数。

```python
from nltk.corpus import movie_reviews as mr
import Levenshtein
import math
from nltk import FreqDist
from nltk.collocations import BigramCollocationFinder
```

```python
#读取所有数据
pathV ="D:/ECNU2021/自然语言处理/作业4/vocab.txt"
pathTD = "D:/ECNU2021/自然语言处理/作业4/testdata.txt"
pathSE = "D:/ECNU2021/自然语言处理/作业4/spell-errors.txt"
vocab = []
testdata = []
spellerrors = {}
with open(pathV,'r') as f:
    while True:
        temp = f.readline()
        if not temp:
            break
        vocab.append(temp[0:-1])
with open(pathTD,'r') as f:
    while True:
        temp = f.readline()
        if not temp:
            break
        spl = temp.strip().split('\t')
        t = spl[2][0:-1].split(' ')
        try:
            while True:
                t.remove(' ')
        except:
            pass
        new_t = []
        #分割时可能会把逗号分进去
        for words in t:
            if len(words)>=2 and words[-1]==',':
                words = words[0:-1]
            new_t.append(words)
        testdata.append((spl[0],spl[1],new_t))
#spellerror中有一些是*2，*3，这里要把它去掉并直接转换成元素以便统计
with open(pathSE,'r') as f:
    while True:
        temp = f.readline()
        if not temp:
            break
        spl=temp.strip().split(':')
        words = spl[1].strip().split(',')
        new_words = []
        for w in words:
            tempTime = 1
            if w[0]==' ':
                w = w[1:]
            if len(w)>2 and w[-2]=='*':
                tempTime = int(w[-1])
                w=w[0:-2]
            for i in range(tempTime):
                new_words.append(w)
        spellerrors[spl[0]]=new_words
#print(spellerrors)
```

```python
#生成候选词集合函数，利用Levenshtein库
def candidate(word):
    ret = []
    for w in vocab:
        if Levenshtein.distance(word,w)<=2:
            ret.append(w)
    return ret
```

Bigram的建立

```python
#构建bigram语言模型
f = mr.fileids()
bigram = []
for dat in f:
    bigram.append(mr.words(dat))
co = []
counter = set()
for dat in bigram:
    for i in dat:
        co.append(i)
        counter.add(i)
count = len(counter)
#co是语料库中所有句子连成的一个list.
#count记录了所有出现过的词汇
```

```python
freq = FreqDist(co)
#freq是单词出现的次数，用于计算p(w)
bigram_count = {}
#bigram_count格式：["word1 word2":freq]
for i in range(0, len(co) - 1):
    bigram = co[i:i + 2]
    bigram = ' '.join(bigram)
    if bigram in bigram_count:
        bigram_count[bigram] += 1
    else:
        bigram_count[bigram] = 1
```

对候选词概率的计算：

1.拼错的概率：p(Right|Wrong)=p(Wrong|Right)\*p(Right)取决于error中w项的数目与是否包含该错词。

2.上下文概率：word1 word word2 p(word|word1)+p(word2|word)

平滑：p(word|word1)=(bigram_count["word1 word"]+1)/(freq["word1"] + V)

这里需要考虑的是，如果这Vocab中的词根本没出现在error库中，这时候说明可能性反而是更小的，这里就需要拟定一个返回值了。观察数据后尝试返回-20。

再次观察后发现，对于Levenshtein距离的计算并没有在概率中体现，因为样本不够多，还不足以计算，因此在这里也要进行加权，因为编辑距离更短的错误更容易发生。由于x|w没有统计数据，因此目前的想法是简单地编辑次数乘以权重5来计算（显然细分编辑的过程更为合理，但是难以做到统计概率）。

```python
V = count
sigma = 5

def prob_w_r(right): #p(Wrong|Right)
    try:
        leng = len(spellerrors[right])
        return math.log(1.0/leng)
    except:
        return -20

def prob_r(right):
    temp = freq.freq(right)
    if temp == 0:
        return -20
    else:
        return math.log(freq.freq(right))

def prob_bi(word1,word2):
    temp = word1 + " " + word2
    if temp in bigram_count:
        cnt = bigram_count[temp]
    else:
        cnt = 0
    prob = (cnt + 1.0)/(freq[word2]+ V)
    return math.log(prob)

def prob_change(wrong,right):
    dist = Levenshtein.distance(wrong,right)
    if dist == 1:
        return sigma
    else:
        return 2*sigma
```

这里有一个问题，如果是数字是没有纠错的必要的，可以用一个函数把数字剔除出去。


```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        pass
    try:
        import unicodedata
        unicodedata.numeric(s)
        return True
    except (TypeError, ValueError):
        pass
    return False
```

尝试对testdata进行纠错,得到纠错结果

```python
TotalMistakes = 0
for data in testdata:
    print("第%d句："% int(data[0]))
    for i in range(len(data[2])):
        word = data[2][i]
        if i !=0:
            pre_word = data[2][i-1]
        else:
            pre_word = '.'
        if i != len(data[2])-1:
            next_word = data[2][i+1]
        else:
            next_word = '.'
        if word not in vocab and len(word)!=0 and not is_number(word):
            TotalMistakes +=1
            candi = candidate(word)
            if len(candi) == 0:
                print(word+" False")
                break
            candi_prob = {}
            for can in candi:
                candi_prob[can] = prob_w_r(can)+prob_r(can)+prob_bi(can,pre_word)+prob_bi(next_word,can)+prob_change(word,can)
            chosen = max(candi_prob,key = candi_prob.get)
            proba = candi_prob[chosen]
            print(word+", "+chosen)
            print(candi_prob)
print("Total Mistakes: %d" % TotalMistakes)
```

总结：

感觉这次实验模块很多，难度不小，其中似然概率计算部分挺头疼的，并且需要注意文本中的标点符号对纠错过程判断的影响。纯数字也是没有纠错价值的。对于p(w|r)的理解还是有一些模糊，计算上对于编辑距离的概率也难以判断，有些遗憾。

最后观察结果，对于单词长度短的词汇来说纠错是非常不容易的，很容易曲解。因此考虑改进方式为词汇长度小于某个数值时只考虑编辑距离为1之内的值进行纠错，可能会有更好的效果。
